## 🔧 How It Works  
1. **User Inputs**: Collects travel preferences via simple prompts  
2. **Gemini AI**: Generates detailed daily itineraries  
3. **Smart Formatting**: Automatically adds emojis and sections  
4. **Save Feature**: Exports plans as `.txt` files 

## 💡 Key Innovations  
✔️ **No boring formats** - Emoji-rich visual planning  
✔️ **Budget-aware** - Cost breakdowns for every activity  
✔️ **Local secrets** - Insider tips from AI  
✔️ **Zero setup** - Runs entirely on Kaggle  

```python
## 1. Setup & Configuration
import google.generativeai as genai
from kaggle_secrets import UserSecretsClient
import re

# Securely load API key
user_secrets = UserSecretsClient()
GOOGLE_API_KEY = user_secrets.get_secret("GOOGLE_API_KEY")
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize model
model = genai.GenerativeModel('gemini-2.0-flash')

# NEW: Emoji-based formatting
def format_itinerary(text: str) -> str:
    """
    Converts raw itinerary into emoji-rich text without * or **.
    """
    # Remove all asterisks
    text = text.replace('*', '')
    
    # Add emojis to sections
    text = re.sub(r'Day \d+:', lambda m: f"\n✨ {m.group(0)} ✨", text)
    text = re.sub(r'(Morning|Afternoon|Evening):', r'🕘 \1:', text)
    text = re.sub(r'Insider Tip:', '💡 Insider Tip:', text)
    text = re.sub(r'Estimated Daily Budget:', '💰 Estimated Daily Budget:', text)
    text = re.sub(r'Pro Tip:', '🌟 Pro Tip:', text)
    
    # Format costs
    text = re.sub(r'(\$[\d,]+)', r'💲 \1', text)
    
    # Clean bullets (use proper hyphens)
    text = re.sub(r'(?<=\n)\s*[-*+]', '  -', text)
    
    return text.strip()

## 2. Core Travel Planner Function (Emoji Edition)
def generate_itinerary(location: str, budget: str, days: int, interests: str) -> str:
    prompt = f"""
    Create a {days}-day itinerary for {location} (Budget: {budget}).
    Focus on: {interests}. Structure like this:

    Day 1: [LOCATION SUMMARY]
    - Morning: [Activity] (Cost: $XX)
      Insider Tip: [Tip]
    - Afternoon: [Activity] (Cost: $XX)
    - Evening: [Activity] (Cost: $XX)

    Estimated Daily Budget: $XX
    Pro Tip: [General advice]

    Use plain text (no * or **). Separate sections clearly.
    """
    try:
        response = model.generate_content(prompt)
        return format_itinerary(response.text)
    except Exception as e:
        return f"❌ Error: {str(e)}"

## 3. User Interface (Emoji Style)
def main():
    print("\n" + "="*50)
    print("🌍 AI TRAVEL PLANNER (EMOJI EDITION) ".center(50))
    print("="*50 + "\n")

    #the inpu() syntax is causing alot of problem in kaggle
    # Get user input
    # location = input("📍 Destination (e.g., Tokyo, Rome): ").strip()
    # budget = input("💰 Budget (e.g., $2000): ").strip()
    # days = input("📅 Days: ").strip()
    # interests = input("🎯 Interests (e.g., food, hiking): ").strip()
    
    # if not all([location, budget, days, interests]):
    #     print("⛔ Please fill all fields!")
    #     return

    
    location="Douala"
    budget="$10000"
    days="3"
    interests="food, culture"

    print("Sample of Itinerary")
    print("📍 Destination: ",location)
    print("💰 Budget: ",budget)
    print("📅 Days: ",days)
    print("🎯 Interests: ",interests)
    
    print("\n" + "✈️ Generating your itinerary...".center(50) + "\n")
    
    # Generate and display
    itinerary = generate_itinerary(location, budget, days, interests)
    print(itinerary)
    
    # Save to file
    with open(f"{location}_itinerary.txt", "w") as f:
        f.write(itinerary)
    print(f"\n✅ Saved as '{location}_itinerary.txt'")

if __name__ == "__main__":
    main()
```

```python
!pip install -q google-generativeai gradio  # Install Gradio

import google.generativeai as genai
import gradio as gr
import os

# Configure Gemini (use your API key)
GOOGLE_API_KEY = "AIzaSyAMo9bhPnoeMLEQoCJeZC9wyGGt_raYNOw"  # Replace or use Kaggle Secrets
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-2.0-flash')

def travel_planner(location, budget, days, interests):
    prompt = f"""
    Create a detailed {days}-day itinerary for {location} with a budget of {budget}. 
    Focus on: {interests}. Include:
    - Daily schedule (morning/afternoon/evening)
    - Cost breakdown (food, attractions, transport)
    - Pro tips (best times to visit, local customs)
    Format with clear headings and bullet points.
    """
    response = model.generate_content(prompt)
    return response.text

# Gradio UI
demo = gr.Interface(
    fn=travel_planner,
    inputs=[
        gr.Textbox(label="📍 Destination", placeholder="e.g., Tokyo, Rome"),
        gr.Textbox(label="💰 Budget", placeholder="e.g., $1000"),
        gr.Textbox(label="📅 Days", placeholder="e.g., 5"),
        gr.Textbox(label="🎯 Interests", placeholder="e.g., museums, hiking")
    ],
    outputs=gr.Textbox(label="✨ Your Personalized Itinerary"),
    title="🌍 AI Travel Planner (Powered by Gemini)",
    description="Generate custom travel plans in seconds!",
    theme="soft"  # Try "default" or "huggingface" for other styles
)

demo.launch(share=True)  # Creates a public link
```
